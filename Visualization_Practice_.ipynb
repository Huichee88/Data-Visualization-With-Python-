###Scatter plot 

Scatter plot can be used for Numerical Grouped by Catergorical variable as well. 

```python
#scatter plot with Pclass against Fare. 
plt.scatter(df['Pclass'], df['Fare'])

#add title 
plt.title("Fare VS Pclass")

#setting X and Y 
plt.ylabel('Fare')
plt.xlabel('Pclass')

plt.show()
```

```python
#plt sex against age
plt.scatter(df['Sex'], df['Age'])

#title
plt.title('Age Vs Sex')

#Xlabel and Ylabel
plt.xlabel('Sex')
plt.ylabel('Age')

plt.colorbar()  #colorbar is the right handside that one. 
plt.show()
```

NOW WE ADD COLORRRR TO THE SCATTER PLOT. 

###Line Chart

Line Chart can be used for Numerical vs Numerical. 

BUT NOT Numerical vs Categorical 

```python
#scatter plot with Age against Fare       #THIS ONE IS NUMERICAL VS NUMERICAL  (CAN/ CORRECT)
plt.plot(df['Fare'])
plt.plot(df['Age'])

#Title
plt.title('Fare VS Age')

#setting labels X and Y
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

```python
#scatterplot of fare against Pclass   #this is numerical vs cateegorical (CANNOT)

plt.plot(df['Fare'])
plt.plot(df['Pclass'])

#title
plt.title('Fare vs Pclass')

#labels
plt.ylabel('Fare')
plt.xlabel('Pclass')

plt.show()
```

###Scatter Plot (SNS)
